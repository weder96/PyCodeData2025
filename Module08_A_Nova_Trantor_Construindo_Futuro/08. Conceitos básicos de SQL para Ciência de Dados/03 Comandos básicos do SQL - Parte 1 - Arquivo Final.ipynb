### O sqlite3

- Vamos tratar uma base fictícia de alunos da Hashtag
- A base já está disponível em:

```python
# Criando a conexão
import sqlite3
con = sqlite3.connect('BaseDados.db')
```

```python
# Criando um cursor
cur = con.cursor()
```

```python
# Buscando todas as informações da tabela
cur.execute('SELECT * FROM dados')
resultado = cur.fetchall()
```

```python
# Salvando em um DataFrame do pandas
import pandas as pd
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# Visualizando os resultados
resultado.tail()
```

```python
# Obtendo o tamanho da base
resultado.shape
```

### Selecionando dados

```python
# Selecionando todos os dados da tabela
cur.execute('SELECT * FROM dados')
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# Selecionando apenas as colunas nome_aluno e cod_matricula
cur.execute('SELECT nome_aluno,cod_matricula FROM dados')
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# Selecionando apenas as colunas nome_aluno e e-mail
comando = 'SELECT nome_aluno,[e-mail] as email FROM dados'
cur.execute(comando)
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# Criando uma função para consultar os dados
def executa_sql(comando):
    cur.execute(comando)
    resultado = cur.fetchall()
    resultado = pd.DataFrame(resultado)
    resultado.columns = [i[0] for i in cur.description]
    print(resultado.shape)
    display(resultado.head())
    return resultado
```

```python
resultado = executa_sql('SELECT nome_aluno,[e-mail] as email FROM dados')
```
