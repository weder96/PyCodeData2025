**Laço de repetição WHILE**
O laço de repetição while é uma estrutura de repetição que continua executando um bloco de código enquanto a condição definida for verdadeira.

Assim que a condição se torna falsa, o loop para automaticamente.

```python
#Sintaxe do comando while
while condição: #Enquanto a condição for verdadeira
    # bloco de código a ser repetido
```

```python
#Exemplo de utilização com contador de 1 até 10
contador = 1

while contador <= 10:
    print("Contagem:", contador)
    contador += 1
```

```python
#Loop infinito
x = 1

while x > 0:
    print(x)
```

```python
#Validação de strings com while
senha = ""

while senha != "python":
    senha = input("Digite a senha: ")

print("Acesso liberado!")
```

```python
#Usuário tem que digitar um valor entre 1 e 10

numero = int(input('Digite um número entre 1 e 10: '))

while numero < 1 or numero > 10:
  numero = int(input('Número inválido. Deve estar entre 1 e 10: '))

print(numero)
```
