```python
import pandas as pd
import numpy as np

def gerar_e_salvar_cotacoes_bancos(nome_arquivo: str):
    """
    Gera dados fictícios de cotações para bancos brasileiros e salva em um arquivo CSV.

    Args:
        nome_arquivo (str): O nome do arquivo CSV a ser criado (ex: 'cotacoes.csv').
    """
    # Passo 1: Definir os tickers e nomes dos bancos
    bancos = {
        'TICKER': ['ITUB4', 'BBDC4', 'BBAS3', 'SANB11', 'BPAC11', 'ITSA4', 'BPAN4', 'ABCB4'],
        'BANCO': ['Itaú Unibanco', 'Bradesco', 'Banco do Brasil', 'Santander Brasil', 
                  'BTG Pactual', 'Itaúsa', 'Banco Pan', 'Banco ABC Brasil']
    }
    num_bancos = len(bancos['TICKER'])

    # Passo 2: Gerar dados numéricos fictícios com NumPy
    # Preços entre 10 e 60
    np.random.seed(42)
    precos_abertura = np.random.uniform(10, 60, num_bancos)
    # Criar alguma variação para os outros preços
    precos_minima = precos_abertura - np.random.uniform(0, 0.5, num_bancos)
    precos_maxima = precos_abertura + np.random.uniform(0, 1, num_bancos)
    precos_fechamento = (precos_minima + precos_maxima) / 2
    # Volume de negociação
    volumes = np.random.randint(5_000_000, 80_000_000, num_bancos)

    # Passo 3: Montar o DataFrame do Pandas
    df_cotacoes = pd.DataFrame({
        'DATA': pd.to_datetime('today').strftime('%Y-%m-%d'),
        'TICKER': bancos['TICKER'],
        'BANCO': bancos['BANCO'],
        'ABERTURA': precos_abertura,
        'MAXIMA': precos_maxima,
        'MINIMA': precos_minima,
        'FECHAMENTO': precos_fechamento,
        'VOLUME': volumes
    })

    # Passo 4: Usar o .to_csv() com parâmetros específicos
    print(f"Gerando o arquivo '{nome_arquivo}'...")
    
    df_cotacoes.to_csv(
        nome_arquivo,
        sep=';',
        index=False,
        decimal=',',
        encoding='utf-8-sig',
        float_format='%.2f'
    )
    
    print(f"Arquivo '{nome_arquivo}' salvo com sucesso!")
    
    return df_cotacoes
```

```python
# --- Exemplo de como usar a função ---
if __name__ == '__main__':
    df_gerado = gerar_e_salvar_cotacoes_bancos('cotacoes_bancos_b3.csv')
    print("\nPré-visualização do DataFrame gerado:")
    print(df_gerado.head())
```
