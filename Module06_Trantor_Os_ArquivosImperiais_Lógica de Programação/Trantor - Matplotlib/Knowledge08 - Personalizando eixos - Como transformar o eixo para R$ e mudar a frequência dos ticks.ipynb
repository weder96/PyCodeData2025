# Knowledge 8 - Como personalizar os eixos do gr√°fico (R$, %)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
#colocando o eixo Y em reais e na cor branca

cotacoes = yf.download(["WEGE3.SA"])['Close']

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')
ax.yaxis.set_major_formatter('R${x:1.0f}')
```

```python
#Mudando a ocorr√™ncia das datas no eixo X

import matplotlib.dates as mdate

cotacoes = yf.download(["WEGE3.SA"], "2010-01-01")['Close']

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)
ax.xaxis.set_major_locator(mdate.YearLocator(2))
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')
```

```python
#Mudando o formato das datas no eixo X

cotacoes = yf.download(["WEGE3.SA"], "2019-01-01")['Close']

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)
minha_formatacao = mdate.DateFormatter('%b-%y')
ax.xaxis.set_major_formatter(minha_formatacao)

plt.xticks(rotation=45)
ax.xaxis.set_major_locator(mdate.MonthLocator(interval = 6))
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')

plt.show()
```

```python
#colocando o eixo em %

import matplotlib.ticker as mtick

cotacoes = yf.download(["WEGE3.SA"], "2015-01-01")['Close']

retornos = cotacoes.pct_change().dropna()

retornos_acumulados = (1 + retornos).cumprod() - 1

fig, ax = plt.subplots()

ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')
ax.plot(retornos_acumulados.index, retornos_acumulados.values)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
```

```python
#mudando a frequencia dos ticks que n√£o s√£o datas

cotacoes = yf.download(["WEGE3.SA"], "2015-01-01")['Close']

retornos = cotacoes.pct_change().dropna()

retornos_acumulados = (1 + retornos).cumprod() - 1

dias = range(0, len(retornos_acumulados))

fig, ax = plt.subplots()

ax.plot(dias, retornos_acumulados.values)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.tick_params(axis='x', colors='w')
ax.tick_params(axis='y', colors='w')

ax.xaxis.set_ticks([0, 1000, 2000], ["Inicio", "Meio", "Fim"])

plt.show()
```
