**Laço de Repetição FOR**
O for é uma estrutura de repetição usada para percorrer uma sequência de dados como listas, strings, intervalos numéricos, entre outros.

Com ele, você pode acessar cada item da sequência, um por vez, e executar uma ação para cada um deles.

```python
#Sintaxe do comando for
'''
for variavel in sequencia:
    Executa bloco de comandos
'''
```

```python
#Exemplo de for com strings
for x in "Python para Dados":
  print(x)
```

```python
#Exemplo de for com listas
nomes = ['Joana', 'Osvaldo', 'Murilo', 'Letícia']

for i in nomes:
  print(i)
```

```python
#For in range - sintaxe

'''
1) for i in range(quantidade de repetições)
2) for i in range(1, 10)
   range(valor inicial, valor final - 1)
3) for i in range(1, 20, 2)
   range(valor inicial, valor final - 1, passo)
'''
```

```python
#For com else
#Imprime uma mensagem ao término do loop

for i in range(1, 10):
  print(i)
else:
  print('Fim da execução do for')
```
