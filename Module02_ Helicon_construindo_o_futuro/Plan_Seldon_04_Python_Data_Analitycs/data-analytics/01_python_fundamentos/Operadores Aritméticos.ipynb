**Operadores Aritméticos**
Os operadores aritméticos são utilizados para realizar cálculos matemáticos entre números (Que podem estar dentro de variáveis)

Em python utilizados os seguintes operadores:
- Soma
- Subtração
- Multiplicação
- Divisão
- Divisão Inteira
- Potência
- Resto da Divisão

```python
#Operador Soma
a = 10
b = 5
print ('Soma: ', a + b)
```

```python
#Operador Subtração
a = 12
b = 3
print ('Subtração: ', a - b)
```

```python
#Operador multiplicação
a = 6
b = 4
print ('Subtração: ', a * b)
```

```python
#Operador divisão
a = 16
b = 4
print ('Divisão: ', a / b)
```

```python
#Operador divisão inteira
a = 9
b = 4
print('Divisão Normal: ', a / b)
print('Divisão Inteira: ', a // b)
```

```python
#Operador potência
a = 2
b = 3
print('Potência: ', a ** b)
```

```python
#Operador Resto da divisão ou módulo
a = 10
b = 3
print('Resto da divisão: ', a % b)
```
