**Listas**
Uma lista é uma estrutura de dados que pode ser utilizada para armazenar  vários elementos, de forma ordenada, utilizando um índice como referência. Ela pode receber valores numéricos do tipo int, float, strings, e estruturas de dados como listas, dicionários....

```python
#Criando listas
nomes = ['João', 'Maria', 'Joaquim']
numeros = [10, 30, 125]
alturas = [1.75, 1.69, 1.54]

x = ['José', 30, 1.98, True]

y = ['Laranja', ['limão', 'morango']]
```

```python
#Acessando item de uma lista
print(nomes[0])
print(nomes[1])
print(nomes[2])
print(nomes[-1])
```

```python
#Acessando um índice inexistente
print(nomes[10])
```

```python
#Verificando a existência de um item na lista
if 'João' in nomes:
  print('O nome João está na lista de nomes')
```

```python
#Modificando um valor dentro de uma lista
frutas = ['Laranja', 'Abacaxi', 'Morango']
print(frutas)

frutas[1] = 'Maracujá'
print(frutas)
```
