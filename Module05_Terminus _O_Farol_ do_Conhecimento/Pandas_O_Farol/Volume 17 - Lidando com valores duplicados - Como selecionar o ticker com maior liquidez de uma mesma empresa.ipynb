```python
import pandas as pd

# 1. Criando nosso DataFrame de operações na bolsa
operacoes = {
    'ticker': ['PETR4', 'VALE3', 'ITUB4', 'PETR4', 'VALE3', 'BBDC4', 'ITUB4', 'PETR4'],
    'tipo_ordem': ['Compra', 'Compra', 'Venda', 'Venda', 'Compra', 'Compra', 'Compra', 'Compra'],
    'corretora': ['XP', 'Clear', 'Rico', 'XP', 'Clear', 'XP', 'rico', 'Clear']
}
df = pd.DataFrame(operacoes)

print("--- DataFrame de Operações ---")
print(df)

```

```python
# Exemplo 1: Quais tickers únicos temos em nossa carteira?
# Usando a função principal do pandas
tickers_unicos = pd.unique(df['ticker'])
print("\n--- 1. Tickers únicos na carteira ---")
print(tickers_unicos)
print(type(tickers_unicos)) # Note que o retorno é um array do NumPy

# Exemplo 2: Quais os tipos de ordem existentes?
# Usando o método direto da Série (mais comum)
tipos_de_ordem = df['tipo_ordem'].unique()
print("\n--- 2. Tipos de ordem únicos ---")
print(tipos_de_ordem)

# Exemplo 3: Verificando as corretoras e encontrando inconsistências
# Aqui o .unique() revela um problema de qualidade nos dados!
corretoras_unicas = df['corretora'].unique()
print("\n--- 3. Corretoras únicas (repare no problema!) ---")
print(corretoras_unicas)
# O resultado ['XP', 'Clear', 'Rico', 'rico'] mostra que 'Rico' e 'rico'
# precisam ser padronizados para não serem contados como duas corretoras diferentes.

# Dica Bônus: E se eu quiser saber apenas a QUANTIDADE de únicos?
# Para isso, use a função .nunique()
num_tickers = df['ticker'].nunique()
print(f"\n--- Bônus: Temos {num_tickers} tickers diferentes na carteira. ---")
```
