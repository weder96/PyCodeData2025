```python
# Criando a conexão
import sqlite3
con = sqlite3.connect('BaseDados.db')
```

```python
# Criando um cursor
cur = con.cursor()
```

```python
# Buscando todas as informações da tabela
import pandas as pd
comando = 'SELECT * FROM dados'
cur.execute(comando)
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# Criando a função para buscar esses dados
def executa_sql(comando): 
    cur.execute(comando)
    resultado = cur.fetchall()
    resultado = pd.DataFrame(resultado)
    if resultado.shape[1] > 0:
        resultado.columns = [i[0] for i in cur.description]
    print(resultado.shape)
    display(resultado.head())
    return resultado
```

```python
resultado_sql = executa_sql('SELECT * FROM dados')
```

```python
# Filtrando colunas
resultado_sql = executa_sql('SELECT nome_aluno, cod_matricula FROM dados')
```

```python
# Filtrando apenas os alunos que FIZERAM a 4 prova
resultado_sql = executa_sql('SELECT nome_aluno, cod_matricula, nr_prova, prova_feita, nota_prova \
                            FROM dados WHERE nr_prova = 4 AND prova_feita = 1')
```

```python
# Filtrando alunos que fizeram a 3 ou 4 prova
resultado_sql = executa_sql('SELECT nome_aluno, cod_matricula, nr_prova, prova_feita, nota_prova \
                            FROM dados WHERE (nr_prova = 4 OR nr_prova = 3) AND prova_feita = 1')
```

```python
display(resultado_sql)
```

```python
# Filtrando os alunos com nota maior que 5 na terceira prova
resultado_sql = executa_sql('SELECT nome_aluno, cod_matricula, nr_prova, prova_feita, nota_prova \
                            FROM dados WHERE nr_prova = 3 AND nota_prova > 5')
```

```python
# Algum aluno tirou menos de 5 na terceira prova?
resultado_sql = executa_sql('SELECT nome_aluno, cod_matricula, nr_prova, prova_feita, nota_prova \
                            FROM dados WHERE nr_prova = 3 AND NOT nota_prova > 5')
```
