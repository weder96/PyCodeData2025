# Aula 4.6 – Visualizando Variáveis com Histogramas e Boxplots


## 🧠 Por que visualizar a distribuição dos dados?

- Ver como os valores estão distribuídos
- Identificar picos, caudas e assimetrias
- Detectar outliers
- Escolher as melhores métricas (média, mediana)

Vamos aprender com dois gráficos muito úteis:
- 📊 **Histograma** – mostra frequência de valores
- 📦 **Boxplot** – mostra dispersão, mediana e outliers


## 🧪 Criando dados simulados

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# Dados normais
dados_normais = np.random.normal(loc=50, scale=10, size=1000)

# Dados com outliers
dados_com_outliers = np.append(dados_normais, [150, 160, 170])

# Define o número de amostras
n = 1000

# Distribuição normal padrão (média = 0, desvio padrão = 1)
normal_data = np.random.normal(loc=1, scale=1, size=n)

# Distribuição com cauda longa (log-normal)
long_tail_data = np.random.lognormal(mean=1, sigma=1.5, size=n)

# Criação do DataFrame
df = pd.DataFrame({
    'normal': normal_data,
    'long_tail': long_tail_data
})

df.head()
```

# 📊 Visualizando com Histogramas


Um histograma é um gráfico de barras que mostra como os valores de uma variável estão distribuídos.

Parece com um gráfico de barras comum, mas tem um detalhe importante:

> Ele agrupa os valores em faixas (ou intervalos) — chamadas de “bins” — e conta quantos dados caem em cada faixa.



**🎯 Exemplo:**

Imagine que você tem as idades de 100 pessoas.

Essas idades vão de 0 a 100 anos. Agora, você quer ver quantas pessoas estão em cada faixa etária:

De 0 a 10 anos

De 10 a 20 anos

De 20 a 30 anos

...

O histograma vai fazer exatamente isso: ele conta quantas idades estão em cada grupo e mostra isso como barras.

**👀 Como se parece um histograma?**

Suponha que você tenha esses dados de idade:

    [18, 19, 20, 22, 23, 24, 30, 31, 33, 37, 38, 39, 40, 42, 45, 50, 70, 80]

E você quer agrupá-los em faixas de 10 em 10 anos.

output (2).png


Cada barra mostra quantos valores existem naquela faixa de idade.

**🧠 E para que serve?**

* Ver onde os dados estão concentrados
* Entender o formato da distribuição (normal, enviesada, etc.)
* Identificar lacunas ou picos incomuns
* Detectar valores fora do padrão (outliers) quando há caudas longas

```python
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribuição Normal')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.hist(dados_com_outliers, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribuição Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.hist(long_tail_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribuição Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python
pd.DataFrame(long_tail_data).median()
```

```python
pd.DataFrame(long_tail_data).mean()
```

# 📦 Lendo um Boxplot

```python
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribuição Normal')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python

import seaborn as sns

sns.boxplot(x=dados_normais, color='lightgreen')
plt.title('Boxplot da Distribuição Normal')
plt.xlabel('Valor')
plt.show()
```

```python
pd.DataFrame(dados_normais).describe()
```

## 🚨 Boxplot com Outliers

```python
plt.hist(dados_com_outliers, bins=30, color='salmon', edgecolor='black')
plt.title('Histograma de Distribuição Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python

sns.boxplot(x=dados_com_outliers, color='salmon')
plt.title('Boxplot com Outliers')
plt.xlabel('Valor')
plt.show()
```

## 🔍 Comparando com e sem outliers

```python

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Sem Outliers')

plt.subplot(1, 2, 2)
plt.hist(dados_com_outliers, bins=30, color='salmon', edgecolor='black')
plt.title('Com Outliers')

plt.tight_layout()
plt.show()
```
