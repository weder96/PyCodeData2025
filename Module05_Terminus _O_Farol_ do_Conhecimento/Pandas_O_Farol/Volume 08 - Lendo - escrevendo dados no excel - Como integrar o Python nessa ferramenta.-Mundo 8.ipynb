```python
import pandas as pd

# --- Dados para a Planilha "Cotacoes" ---
dados_cotacoes = {
    'Data': [
        '2025-09-22', '2025-09-22', '2025-09-23', '2025-09-23',
        '2025-09-24', '2025-09-24', '2025-09-25', '2025-09-25'
    ],
    'Ticker': [
        'ITUB4', 'BBDC4', 'ITUB4', 'BBDC4',
        'BBAS3', 'SANB11', 'BBAS3', 'SANB11'
    ],
    'Abertura': [32.50, 15.20, 32.85, 15.05, 55.80, 30.10, 56.00, 29.90],
    'Fechamento': [32.80, 15.10, 33.10, 15.25, 56.10, 30.00, 56.50, 29.80],
    'Volume': [50000000, 45000000, 52000000, 48000000, 60000000, 30000000, 62000000, 31000000]
}
df_cotacoes = pd.DataFrame(dados_cotacoes)
# Convertendo a coluna 'Data' para o tipo datetime, para o Excel formatar corretamente
df_cotacoes['Data'] = pd.to_datetime(df_cotacoes['Data'])


# --- Dados para a Planilha "Proventos" ---
dados_proventos = {
    'Ticker': ['ITUB4', 'BBDC4', 'BBAS3', 'SANB11'],
    'Tipo_Provento': ['Dividendo', 'JCP', 'Dividendo', 'JCP'],
    'Data_Com': ['2025-08-15', '2025-08-20', '2025-09-01', '2025-09-05'],
    'Valor_por_Acao': [0.75, 0.45, 1.10, 0.60]
}
df_proventos = pd.DataFrame(dados_proventos)
df_proventos['Data_Com'] = pd.to_datetime(df_proventos['Data_Com'])


# --- Escrevendo os DataFrames no arquivo Excel ---
nome_arquivo = 'dados_bancos_b3.xlsx'

# Usamos o ExcelWriter para conseguir escrever em múltiplas abas (planilhas)
with pd.ExcelWriter(nome_arquivo, engine='openpyxl') as writer:
    # Escrevendo a planilha "Cotacoes"
    # O 'index=False' evita que o índice do DataFrame (0, 1, 2...) seja escrito no arquivo
    df_cotacoes.to_excel(writer, sheet_name='Cotacoes', index=False, startrow=2)
    
    # Escrevendo a planilha "Proventos"
    df_proventos.to_excel(writer, sheet_name='Proventos', index=False)
    
    # Adicionando o título na planilha "Cotacoes" (passo extra para replicar o cenário)
    # Acessamos a planilha que acabamos de criar
    worksheet_cotacoes = writer.sheets['Cotacoes']
    # Escrevemos o título na primeira célula (A1)
    worksheet_cotacoes['A1'] = 'Relatório Diário de Cotações - Setembro 2025'

print(f"Arquivo '{nome_arquivo}' criado com sucesso!")
print("Ele contém as planilhas: 'Cotacoes' e 'Proventos'.")
```
