# Aula 4.6 ‚Äì Visualizando Vari√°veis com Histogramas e Boxplots


## üß† Por que visualizar a distribui√ß√£o dos dados?

- Ver como os valores est√£o distribu√≠dos
- Identificar picos, caudas e assimetrias
- Detectar outliers
- Escolher as melhores m√©tricas (m√©dia, mediana)

Vamos aprender com dois gr√°ficos muito √∫teis:
- üìä **Histograma** ‚Äì mostra frequ√™ncia de valores
- üì¶ **Boxplot** ‚Äì mostra dispers√£o, mediana e outliers


## üß™ Criando dados simulados

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# Dados normais
dados_normais = np.random.normal(loc=50, scale=10, size=1000)

# Dados com outliers
dados_com_outliers = np.append(dados_normais, [150, 160, 170])

# Define o n√∫mero de amostras
n = 1000

# Distribui√ß√£o normal padr√£o (m√©dia = 0, desvio padr√£o = 1)
normal_data = np.random.normal(loc=1, scale=1, size=n)

# Distribui√ß√£o com cauda longa (log-normal)
long_tail_data = np.random.lognormal(mean=1, sigma=1.5, size=n)

# Cria√ß√£o do DataFrame
df = pd.DataFrame({
    'normal': normal_data,
    'long_tail': long_tail_data
})

df.head()
```

# üìä Visualizando com Histogramas


Um histograma √© um gr√°fico de barras que mostra como os valores de uma vari√°vel est√£o distribu√≠dos.

Parece com um gr√°fico de barras comum, mas tem um detalhe importante:

> Ele agrupa os valores em faixas (ou intervalos) ‚Äî chamadas de ‚Äúbins‚Äù ‚Äî e conta quantos dados caem em cada faixa.



**üéØ Exemplo:**

Imagine que voc√™ tem as idades de 100 pessoas.

Essas idades v√£o de 0 a 100 anos. Agora, voc√™ quer ver quantas pessoas est√£o em cada faixa et√°ria:

De 0 a 10 anos

De 10 a 20 anos

De 20 a 30 anos

...

O histograma vai fazer exatamente isso: ele conta quantas idades est√£o em cada grupo e mostra isso como barras.

**üëÄ Como se parece um histograma?**

Suponha que voc√™ tenha esses dados de idade:

    [18, 19, 20, 22, 23, 24, 30, 31, 33, 37, 38, 39, 40, 42, 45, 50, 70, 80]

E voc√™ quer agrup√°-los em faixas de 10 em 10 anos.

output (2).png


Cada barra mostra quantos valores existem naquela faixa de idade.

**üß† E para que serve?**

* Ver onde os dados est√£o concentrados
* Entender o formato da distribui√ß√£o (normal, enviesada, etc.)
* Identificar lacunas ou picos incomuns
* Detectar valores fora do padr√£o (outliers) quando h√° caudas longas

```python
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
plt.hist(dados_com_outliers, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribui√ß√£o Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
plt.hist(long_tail_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribui√ß√£o Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
pd.DataFrame(long_tail_data).median()
```

```python
pd.DataFrame(long_tail_data).mean()
```

# üì¶ Lendo um Boxplot

```python
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python

import seaborn as sns

sns.boxplot(x=dados_normais, color='lightgreen')
plt.title('Boxplot da Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.show()
```

```python
pd.DataFrame(dados_normais).describe()
```

## üö® Boxplot com Outliers

```python
plt.hist(dados_com_outliers, bins=30, color='salmon', edgecolor='black')
plt.title('Histograma de Distribui√ß√£o Normal com outliers')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python

sns.boxplot(x=dados_com_outliers, color='salmon')
plt.title('Boxplot com Outliers')
plt.xlabel('Valor')
plt.show()
```

## üîç Comparando com e sem outliers

```python

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(dados_normais, bins=30, color='skyblue', edgecolor='black')
plt.title('Sem Outliers')

plt.subplot(1, 2, 2)
plt.hist(dados_com_outliers, bins=30, color='salmon', edgecolor='black')
plt.title('Com Outliers')

plt.tight_layout()
plt.show()
```
