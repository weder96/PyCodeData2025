```python
#Percorrendo um dicionário

pessoas = {
    "nome": "Sonia dos Santos",
    "idade": 32,
    "altura": 1.69,
    "habilitado": False
}

print(pessoas)

for chave in pessoas:
  print(chave)


```

```python
#Impressão dos valores do dicionário
for chave in pessoas:
  print(pessoas[chave])
```

```python
#Impressão chave: valor do dicionário
for chave in pessoas:
  print(f"{chave} : {pessoas[chave]}")
```

```python
#Utilizando método keys()

for chave in pessoas.keys():
  print(chave, end=" ")

print()

for chave in pessoas.keys():
  print(pessoas[chave], end=" ")
```

```python
#Utilizando método values()

for valores in pessoas.values():
  print(valores)
```

```python
#Utilizando método items()

for chave, valor in pessoas.items():
  print(f"{chave} : {valor}")
```

```python
#Fazendo uma cópia de um dicionário

produto1 = {
    "id": 1045,
    "nome": "Notebook Dell",
    "preco": 2899.90,
    "estoque": 17,
    "categoria": "Eletrônicos"
}

```

```python
#Fazendo a cópia
produto2 = produto1
```

```python
print(f"produto1 : {produto1}")

print(f"produto2 : {produto2}")
```

```python
produto1["preco"] = 2999.90
```

```python
print(f"produto1 : {produto1}")
```

```python
print(f"produto2 : {produto2}")
```

```python
#Utilizando método copy()

produto2 = produto1.copy()
```
