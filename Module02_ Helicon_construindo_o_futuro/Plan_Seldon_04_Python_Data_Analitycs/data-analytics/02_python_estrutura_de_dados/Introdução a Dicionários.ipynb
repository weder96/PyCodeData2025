**DICIONÁRIOS**

Um dicionário em Python é uma coleção não ordenada e mutável, composta por pares de chave e valor (key: value)

```python
#Sintaxe Básica
pessoas = {
    "nome": 'Joaquim da Silva',
    "idade": 32,
    "altura": 1.88,
    "habilitado": True
}

print(pessoas)
print(type(pessoas))
```

```python
#Acessando um item de um dicionário

print(pessoas['nome'])
print(pessoas['idade'])
print(pessoas['altura'])
print(pessoas['habilitado'])
```

```python
#Não aceita valores duplicados
carro = {
    "nome": 'Gol',
    "marca": 'Volkswagen',
    "modelo": 'G5',
    "modelo": 'G4'
}

print(carro)
```

```python
#Modificando um item de um dicionário

carro = {
    "nome": 'Gol',
    "marca": 'Volkswagen',
    "modelo": 'G5',
}

print(carro["modelo"])
carro["modelo"] = "Trend"

print(carro["modelo"])
```

```python
#Tamanho de um dicionário

print(len(carro))
```
