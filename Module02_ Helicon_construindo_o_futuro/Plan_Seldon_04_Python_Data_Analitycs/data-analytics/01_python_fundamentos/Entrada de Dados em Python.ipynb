```python
#Comando input para entrada de dados em pyhton
nome = input('Digite o seu nome: ')
print(nome)
```

```python
#Lendo um número inteiro
idade = input('Digite a sua idade: ')
print(idade)
```

```python
#Verificando o tipo da variável idade
type(idade)
```

```python
#Lendo e convertendo para inteiro
idade = int(input('Digite a sua idade: '))
print(idade)
type(idade)
```

```python
#Lendo e convertendo para float
altura = float(input('Digite a sua altura: '))
print(altura)
type(altura)
```
