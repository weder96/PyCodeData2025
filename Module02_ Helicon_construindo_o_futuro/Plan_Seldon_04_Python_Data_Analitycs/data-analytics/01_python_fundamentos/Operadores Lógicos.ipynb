**Operadores Lógicos**
Os operadores lógicos são usados para conectar duas ou mais expressões booleanas, ou seja, expressões que retornam True ou False."

O resultado final também será um valor booleano

```python
#Operador E

idade = 25
tem_carteira = True

idade >= 18 and tem_carteira
```

```python
#Operador OU
tem_ingresso = False
nome_na_lista = True

tem_ingresso or nome_na_lista
```

```python
#Operador NÃO
logado = False

not logado
```

```python
#Prioridade dos operadores
#parênteses, not, and e por último or
```

```python
#Verificar o resultado seguindo a prioridade
a = True
b = False
c = True

resultado = not a or b and c
resultado
```
