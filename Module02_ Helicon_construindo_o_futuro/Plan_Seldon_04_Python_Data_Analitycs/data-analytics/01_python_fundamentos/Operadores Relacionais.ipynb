**Operadores Relacionais**
Os Operadores relacionais servem para comparar valores. O resultado de uma comparação sempre será um valor booleano: True (verdadeiro) ou False (falso)."

```python
#Operador de igualdade
a = 10
b = 8
a == b
```

```python
#Maior que
a = 5
b = 3

a > b
b > a
```

```python
#Maior ou igual a
a = 7
b = 5

a >= b
b >= a
```

```python
#Menor que
a = 7
b = 5

a < b
b < a
```

```python
#Menor ou igual a
a = 7
b = 5

a <= b
b <= a
```

```python
#Diferente de
a = 7
b = 5

a != b
```
