```python
import pandas as pd
import numpy as np
tickers = ['ITUB4', 'BBDC4', 'BBAS3', 'SANB11']
datas = pd.date_range(start='2025-09-15', periods=10, freq='B') # 10 dias úteis

np.random.seed(42) # Semente para reprodutibilidade
dados_retornos = np.random.normal(loc=0.05, scale=2, size=(10, 4))
# Volume negociado em R$ (entre 30M e 100M)
dados_volume = np.random.randint(low=30_000_000, high=100_000_000, size=(10, 4))
df_retornos = pd.DataFrame(dados_retornos, index=datas, columns=tickers)
df_volume = pd.DataFrame(dados_volume, index=datas, columns=tickers)

print("--- Retorno Diário Percentual (%) ---")
print(df_retornos.round(2))

print("\n--- Volume Financeiro Negociado (R$) ---")
print(df_volume)
```

```python
# Ordenando o índice em ordem crescente (padrão)
df_ordenado = df_retornos.sort_index()
print("\n--- DataFrame com Índice Ordenado Cronologicamente ---")
print(df_ordenado)
```

```python
# Ordenando do mais recente para o mais antigo
df_recente_primeiro = df_retornos.sort_index(ascending=False)

print("\n--- DataFrame com os Dados Mais Recentes Primeiro ---")
print(df_recente_primeiro)
```

```python
# Ordenando as colunas (tickers) em ordem alfabética
df_colunas_ordenadas = df_retornos.sort_index(axis=1)

print("\n--- DataFrame com Colunas (Tickers) Ordenadas ---")
print(df_colunas_ordenadas)
```

```python
df_para_modificar = df_retornos.copy() 
# Copiando para não alterar o original dos outros exemplos

print("\n--- DataFrame Antes da Modificação Inplace ---")
print(df_para_modificar.index)

# Ordenando o próprio DataFrame, sem criar um novo
df_para_modificar.sort_index(inplace=True)
print("\n--- DataFrame Após a Modificação Inplace ---")
print(df_para_modificar.index)
```

```python
# Ordenando pela coluna 'Preco'
df_ordenado_preco = df_retornos.sort_values(by='ITUB4')

print("\n--- Bancos ordenados por ITUB4 (crescente) ---")
print(df_ordenado_preco)
```

```python
# Ordenando pela coluna 'Volume' em ordem decrescente
df_ordenado_volume = df_retornos.sort_values(by='ITUB4', ascending=False)

print("\n--- Bancos ordenados por Volume (decrescente) ---")
print(df_ordenado_volume)
```

```python
data_analise = '2025-09-16'
retornos_do_dia = df_retornos.loc[data_analise]
volumes_do_dia = df_volume.loc[data_analise]

df_dia = pd.DataFrame({'Retorno_%': retornos_do_dia,'Volume': volumes_do_dia })

print("--- Dados combinados para o dia", data_analise, "---")
print(df_dia)

df_modificado = df_dia.copy()
retorno_para_empate = df_modificado.loc['BBAS3', 'Retorno_%']
df_modificado.loc['ITUB4', 'Retorno_%'] = retorno_para_empate

print("\n--- DataFrame Modificado (com empate forçado) ---")
print(df_modificado.round(2))


# 4. Ordenar primeiro por 'Retorno_%' e depois por 'Volume'
df_multi_sort = df_modificado.sort_values(
    by=['Retorno_%', 'Volume'],
    ascending=[False, False] # Retorno decrescente, Volume decrescente
)

print("\n--- Bancos ordenados por Retorno e Volume como desempate ---")
print(df_multi_sort.round(2))
```

```python
df_com_nan = df_retornos.copy()

# Adicionando um valor nulo (NaN) no retorno de ITUB4 em um dia específico.
# Note que usamos a data '2025-09-17' como rótulo do índice, em vez do número 1.
df_com_nan.loc['2025-09-17', 'ITUB4'] = np.nan
df_com_nan.loc['2025-09-18', 'ITUB4'] = np.nan

print("--- DataFrame com valor Nulo (NaN) em ITUB4 ---")
print(df_com_nan.round(2))

print("\n--- Ordenação por ITUB4 com NaN no final (padrão) ---")
print(df_com_nan.sort_values(by='ITUB4').round(2))


print("\n--- Ordenação por ITUB4 com NaN no início ---")
print(df_com_nan.sort_values(by='ITUB4', na_position='first').round(2))
```

```python
import pandas as pd

# Criando dados fictícios de performance para os bancos
dados = {
    'Ticker': ['ITUB4', 'BBDC4', 'BBAS3', 'SANB11'],
    'Retorno_%': [0.92, -0.66, 0.54, -0.33],
    'Dividend_Yield_%': [4.57, 5.30, 3.74, 4.00],
    'Volatilidade_%': [22.5, 21.8, 28.0, 25.4],
    'Volume_Negociado': [75_000_000, 60_000_000, 90_000_000, 40_000_000]
}
df_bancos = pd.DataFrame(dados).set_index('Ticker')

print("--- DataFrame Original com Métricas dos Bancos ---")
print(df_bancos)
```

```python
# Criando uma nova coluna com o ranking do retorno
df_bancos['Rank_Retorno'] = df_bancos['Retorno_%'].rank()

print("\n--- Ranking de Retorno (Padrão: crescente) ---")
print(df_bancos[['Retorno_%', 'Rank_Retorno']])
```

```python
# Ranking de Dividend Yield, do maior para o menor
df_bancos['Rank_DY'] = df_bancos['Dividend_Yield_%'].rank(ascending=False)

print("\n--- Ranking de Dividend Yield (Decrescente) ---")
print(df_bancos[['Dividend_Yield_%', 'Rank_DY']])
```

```python
df_empate = df_bancos.copy()
df_empate.loc['ITUB4', 'Volatilidade_%'] = 25.4 # Forçando um empate com SANB11

# Testando diferentes métodos para lidar com o empate
df_empate['Rank_Vol_Average'] = df_empate['Volatilidade_%'].rank(method='average') # Padrão
df_empate['Rank_Vol_Min'] = df_empate['Volatilidade_%'].rank(method='min')
df_empate['Rank_Vol_First'] = df_empate['Volatilidade_%'].rank(method='first')

print("\n--- Lidando com Empates na Volatilidade ---")
print(df_empate[['Volatilidade_%', 'Rank_Vol_Average', 'Rank_Vol_Min', 'Rank_Vol_First']])
```

```python
# Calcula o ranking como um percentual
df_bancos['Rank_Volume_Pct'] = df_bancos['Volume_Negociado'].rank(pct=True)

print("\n--- Ranking Percentual do Volume Negociado ---")
print(df_bancos[['Volume_Negociado', 'Rank_Volume_Pct']])
```

```python
#df_base

dicionario = {
       "nomes": ["Weg", "Petrobras", "Vale", "Petrobras", "Lojas Renner"],
       "tickers": ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],  
       "cotacoes": [20, 30, 40, 12, 35],
       "preco_sobre_lucro": [25, 6, 12, 7, 25],  
       "volume": [5000, 1000, 4000, 7000, 1200]
}

df_info_empresas = pd.DataFrame(dicionario)

df_info_empresas = df_info_empresas.set_index("tickers")

df_info_empresas.columns = ['nomes', 'precos', 'preco_sobre_lucro', 'volume']

df_info_empresas['lpa'] = df_info_empresas['precos'] / df_info_empresas['preco_sobre_lucro']

dolar = 5.25

df_info_empresas['precos_dolar'] = df_info_empresas['precos'] / dolar

df_info_empresas
```

```python
#gabarito 53

df_info_empresas = df_info_empresas.sort_index()

df_info_empresas
```

```python
#gabarito 54

df_info_empresas = df_info_empresas.sort_values(by = "preco_sobre_lucro", ascending = False)

df_info_empresas
```

```python
#gabarito 55

df_info_empresas = df_info_empresas.sort_values(by = ["preco_sobre_lucro", "volume"])

df_info_empresas
```

```python
#gabarito 56

df_info_empresas['rank_lpa'] = df_info_empresas['lpa'].rank(method = "max")

df_info_empresas
```
