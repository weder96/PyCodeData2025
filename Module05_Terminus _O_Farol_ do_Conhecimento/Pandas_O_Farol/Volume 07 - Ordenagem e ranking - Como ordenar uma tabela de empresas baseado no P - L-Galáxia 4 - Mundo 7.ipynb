### Introdução ao Pandas

```python
import pandas as pd
import numpy as np
```

```python
def gerador_serie_historica(valor_inicial, volatilidade, periodos, dia_inicial, frequencia = "M"):
    
    vetor = [valor_inicial]

    for i in range(periodos - 1):

            preco = vetor[i] * (1 + np.random.normal(0, volatilidade))

            vetor.append(preco)


    serie = pd.Series(vetor, index = pd.date_range(dia_inicial, 
                                                   periods = periodos, freq = frequencia)) 
                               
    return serie
```

```python
serie_lucro = gerador_serie_historica(2, 0.5, 5, "2022-01-01")
serie_cotacoes = gerador_serie_historica(20, 0.05, 5, "2022-01-01")
df = pd.DataFrame({"cotacao": serie_cotacoes,
                 "lucro_por_acao": serie_lucro})
df['PL'] = df['cotacao']/df['lucro_por_acao']
df
```

```python
dicionario = {
       "nomes": ["Quero Quero", "Alpargatas", "Alpargatas", "Magazine Luiza"],
       "preco_sobre_lucro": [12, 6, 12, 100],  
       "volume": [5000, 1000, 4000, 7000]
}

empresas = pd.DataFrame(dicionario, index = ["LJQQ3", "ALPA3", "ALPA4", "MGLU3"])

empresas
```

```python
#sort index 

#crescente
empresas = empresas.sort_index()

print(empresas)

#decrescente
empresas = empresas.sort_index(ascending = False)

print(empresas)

#por coluna

empresas = empresas.sort_index(axis = 1, ascending = False)

empresas

#axis = 1 é quando queremos aplicar o método a coluna. 
```

```python
df = df.sort_index(ascending = False)
df
```

```python
#sort values para ordernar colunas

df = df.sort_values(by = "PL")
print(df)

df = df.sort_values(by = "PL", ascending = False)
df
```

```python
#ordernar por duas colunas

empresas = empresas.sort_values(by = ["preco_sobre_lucro", "volume"]) 
#sempre utilizando listas quando precisamos de mais de um coluna

empresas
```

```python
#criando rankings nos dataframes

#padrao
empresas['raking_padrao'] = empresas['preco_sobre_lucro'].rank()

empresas
```

```python
#min e max
empresas['raking_max'] = empresas['preco_sobre_lucro'].rank(method = "max")
empresas['raking_min'] = empresas['preco_sobre_lucro'].rank(method = "min")
empresas['raking_dense'] = empresas['preco_sobre_lucro'].rank(method = "dense")

empresas
```

```python
#df_base

dicionario = {
       "nomes": ["Weg", "Petrobras", "Vale", "Petrobras", "Lojas Renner"],
       "tickers": ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],  
       "cotacoes": [20, 30, 40, 12, 35],
       "preco_sobre_lucro": [25, 6, 12, 7, 25],  
       "volume": [5000, 1000, 4000, 7000, 1200]
}

df_info_empresas = pd.DataFrame(dicionario)

df_info_empresas = df_info_empresas.set_index("tickers")

df_info_empresas.columns = ['nomes', 'precos', 'preco_sobre_lucro', 'volume']

df_info_empresas['lpa'] = df_info_empresas['precos'] / df_info_empresas['preco_sobre_lucro']

dolar = 5.25

df_info_empresas['precos_dolar'] = df_info_empresas['precos'] / dolar

df_info_empresas
```

```python
#gabarito 53

df_info_empresas = df_info_empresas.sort_index()

df_info_empresas
```

```python
#gabarito 54

df_info_empresas = df_info_empresas.sort_values(by = "preco_sobre_lucro", ascending = False)

df_info_empresas
```

```python
#gabarito 55

df_info_empresas = df_info_empresas.sort_values(by = ["preco_sobre_lucro", "volume"])

df_info_empresas
```

```python
#gabarito 56

df_info_empresas['rank_lpa'] = df_info_empresas['lpa'].rank(method = "max")

df_info_empresas
```
