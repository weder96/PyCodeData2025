```python
#Percorrendo uma string com for
texto = 'Python para dados'
for i in texto:
  print(i)
```

```python
#Percorrendo uma string com while
texto = 'Python para dados'
i = 0
while i < len(texto):
    print(texto[i])
    i += 1
```

```python
#Fatiamento de Strings
texto = 'Python para dados'
print(texto)

#texto[inicio: fim - 1: passo]
```

```python
print(texto[0])
print(texto[-1])
print(texto[::])
print(texto[0:3:1])
print(texto[::-1])
print(texto[-1:-6:-1])
```
