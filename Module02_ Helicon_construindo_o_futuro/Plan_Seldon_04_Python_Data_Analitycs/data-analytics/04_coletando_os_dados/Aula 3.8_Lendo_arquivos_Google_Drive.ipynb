# Lendo arquivos diretamente do Google Drive no Colab


✅ Ideal para: CSVs, Excel, imagens, e qualquer outro arquivo usado em análise de dados.

## Passo 1 – Conectando o Colab ao seu Google Drive
Isso permite que você navegue pelas suas pastas e leia arquivos diretamente de lá.

```python
from google.colab import drive

drive.mount('/content/drive')
```

### Isso vai abrir uma caixinha pedindo permissão. Clique no link, copie o código e cole no campo indicado.

## Passo 2 – Encontrando o caminho do seu arquivo
Depois de montar o Drive, você verá uma pasta `/content/drive/MyDrive/`.

Coloque seu arquivo na pasta **Meu Drive** ou em uma subpasta e copie o caminho completo.

### Exemplo:
Se seu arquivo estiver em:
`Meu Drive/dados/clientes.csv`
Use:
`/content/drive/MyDrive/dados/clientes.csv`

## Passo 3 – Lendo o arquivo normalmente com pandas

```python
import pandas as pd
```

```python
caminho = '/content/drive/MyDrive/notas/clientes_dataset.xlsx'
df = pd.read_excel(caminho)
df.head()
```
