### Limpeza e tratamento de dados

```python
import pandas as pd
import yfinance as yf
```

```python
dados_acoes = yf.download(["ABEV3.SA", "MGLU3.SA", "WEGE3.SA", "VALE3.SA"])['Close']

dados_acoes
```

```python
media_retornos_diarios = dados_acoes.pct_change().dropna().mean()

media_retornos_diarios
```

```python
#usando map pra transformar dados

dados_anuais = media_retornos_diarios.map(lambda x: (1 + x) ** 252 - 1)

dados_anuais
```

```python
dados_anuais = dados_anuais.to_frame()

dados_anuais
```

```python
dados_anuais = dados_anuais.reset_index()

dados_anuais.columns = ["nome_empresas", "retornos_anuais"]

dados_anuais
```

```python
dados_anuais['nome_empresas'] = dados_anuais['nome_empresas'].map(lambda x: x.lower())

dados_anuais

#sempre que você precisar aplicar uma função a uma coluna, utilize o map :)
```

```python
import pandas as pd
import numpy as np

# 1. Nosso DataFrame inicial com uma coluna de tickers
df = pd.DataFrame({
    'ticker': ['ITUB4', 'BBAS3', 'BBDC4', 'SANB11', 'BPAC11', 'ABC B4']
})
print("--- 1. DataFrame Original ---")
print(df)

# 2. Criamos nossos "mapas" (dicionários) de informação
# É como uma tabela de consulta ou uma legenda
mapa_nomes = {
    'ITUB4': 'Itaú Unibanco',
    'BBAS3': 'Banco do Brasil',
    'BBDC4': 'Bradesco',
    'SANB11': 'Santander Brasil',
    'BPAC11': 'BTG Pactual'
    # Note que 'ABC B4' não está aqui de propósito
}

mapa_segmento = {
    'ITUB4': 'Banco de Varejo',
    'BBAS3': 'Banco de Varejo',
    'BBDC4': 'Banco de Varejo',
    'SANB11': 'Banco de Varejo',
    'BPAC11': 'Banco de Investimento'
}

# 3. Usando .map() para criar uma nova coluna com os nomes das empresas
# A mágica acontece aqui!
print("\n--- 2. Criando a coluna 'nome_banco' com .map() ---")
df['nome_banco'] = df['ticker'].map(mapa_nomes)
print(df)
print("\n>>> Repare que 'ABC B4' virou NaN porque não estava no mapa!")


# 4. Criando outra coluna para o segmento
print("\n--- 3. Criando a coluna 'segmento' ---")
df['segmento'] = df['ticker'].map(mapa_segmento)
print(df)


# 5. Exemplo com uma FUNÇÃO: Classificar o tipo de ação pelo ticker
# '3' = Ordinária, '4' = Preferencial, '11' = Unit
def classificar_acao(ticker):
    if ticker.endswith('3'):
        return 'Ordinária (ON)'
    elif ticker.endswith('4'):
        return 'Preferencial (PN)'
    elif ticker.endswith('11'):
        return 'Unit'
    else:
        return 'Outro'

print("\n--- 4. Usando .map() com uma função para classificar a ação ---")
df['tipo_acao'] = df['ticker'].map(classificar_acao)
print(df)
```

Use o DF base do mundo 17 para:

* Exercício 71: Transforma a coluna de cotações para reais. Ex: 20 -> R$20

```python
#df_base

dicionario = {
       "nomes": ["Weg", "Petrobras", "Vale", "Petrobras", "Lojas Renner"],
       "tickers": ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],  
       "cotacoes": [20, 30, 40, 12, 35],
       "preco_sobre_lucro": [25, 6, 12, 7, 25],  
       "volume": [5000, 1000, 4000, 7000, 1200]
}

df_info_empresas = pd.DataFrame(dicionario)

df_info_empresas = df_info_empresas.set_index("tickers")

df_info_empresas.columns = ['nomes', 'precos', 'preco_sobre_lucro', 'volume']

df_info_empresas['lpa'] = df_info_empresas['precos'] / df_info_empresas['preco_sobre_lucro']

dolar = 5.25

df_info_empresas['precos_dolar'] = df_info_empresas['precos'] / dolar

df_info_empresas
```

```python
#gabarito 70

df_info_empresas['nomes'] = df_info_empresas['nomes'].map(lambda x: x.lower())

df_info_empresas
```

```python
#gabarito 71 

df_info_empresas['precos'] = df_info_empresas['precos'].map(lambda x: "R$ " + str(x))

df_info_empresas
```
