```python
#Instrução while True
while True:
    senha = input("Digite sua senha: ")
    if senha == "python123":
        print("Acesso liberado!")
        break
    else:
        print("Senha incorreta, tente novamente.")
```

```python
#Instrução while True - Leia números e imprima na tela até o usuário digitar o número 0.
while True:
    numeros = int(input('Digite um número inteiro: '))
    if numeros == 0:
        print('Programa encerrado!')
        break
    else:
        print('Número digitado: ', numeros)
```

```python
#Comando break
contador = 0

while True:
    print(contador)
    if contador == 3:
        break
    contador += 1
```

```python
#Comando continue - Imprima de 1 a 20, com exceção do número 10
x = 1

while x < 21:
    if x == 10:
        continue
    print(x)
    x += 1
```
