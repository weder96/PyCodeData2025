```python
import pandas as pd
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = pd.read_csv("../../data/others/anscombe.csv")
#df = sns.load_dataset("anscombe")

#seaborn components used: set_theme(), load_dataset(), lmplot()
# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```

# Anscombe's Quartet: The Lesson Every Data Scientist Learns

You know, in the world of data, there's a classic trap that's incredibly easy to fall into, especially early in your career. It's the temptation to blindly trust the numbersthe summary statistics. And the best example to illustrate this danger is the famous **Anscombe's Quartet**. I love showing this plot because it's the visual proof of why our job is as much an art as it is a science.

What the code you saw does is brilliant in its simplicity. It takes this quartet, which is four seemingly "identical" datasets, and forces us to truly *see* them.

***

### The Illusion of Numbers

First, let's talk about the "gotcha" of the quartet. The statistician Francis Anscombe created four datasets that, if you run a basic statistical analysis (using `df.describe()`, for example), will give you nearly identical results:

* The mean of `x` and `y`? The same across all four.
* The variance? The same.
* The correlation between `x` and `y`? Identical.
* And most impressively: the **linear regression line** (`y = mx + b`) that describes the trend in the data? It's virtually the same for all four!

If we were just looking at a spreadsheet or a statistical report, our conclusion would be obvious: "Okay, these four datasets tell the same story. Next!" And that's precisely where the mistake happens.

***

### The Truth Revealed by the Graph

This is where `seaborn.lmplot` comes in as our hero. We ask it to do two things: plot the data points of each dataset (a scatter plot) and draw that very regression line we calculated (a linear model plot). By using `col="dataset"`, we're telling it, "Hey, Seaborn, show me each of these side-by-side so I can compare them."

The result is one of the most important "Aha!" moments in data analysis.

**Graph Anscombe's Quartet Plot**

When you look at the four plots, you see the truth:

* **Dataset I**: This is our "poster child." A clear, well-behaved linear relationship. The regression line describes the data very well. This is what we hope to find.
* **Dataset II**: Here, we realize instantly: this isn't a line, it's a curve! Using a linear model here is simply the wrong model. The statistics completely deceived us.
* **Dataset III**: We have a nearly perfect line, but with a single **outlier**. This one point was enough to "pull" the regression line down and distort reality.
* **Dataset IV**: This one is perhaps the most treacherous. It's not a linear relationship at all. In fact, without that single **high-leverage point** on the far right, there would be no relationship whatsoever! That lone point is single-handedly dictating the entire regression line.

***

### The Golden Rule

So, the objective of this graph is to teach the golden rule of data science in a way you'll never forget: **ALWAYS plot your data!**

Summary statistics are useful, but they don't tell the whole story. They are the summary of a book, but without reading the chapters, you miss all the context, the nuances, and the main characters (or the outliers!). This graph is a humble reminder that before building complex models, our first and most important job is simply **to look**.
