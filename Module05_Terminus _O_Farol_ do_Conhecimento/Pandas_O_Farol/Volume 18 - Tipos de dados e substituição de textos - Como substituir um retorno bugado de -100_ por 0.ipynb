```python
import pandas as pd

# 1. Criando nosso DataFrame com dados "errados" de ações de bancos da B3
data = {
    'id_pregao': [1001.0, 1002.0, 1003.0, 1004.0, 1005.0],
    'data_pregao': ['2025-10-06', '2025-10-06', '2025-10-07', '2025-10-07', '2025-10-08'],
    'ticker': ['ITUB4', 'BBDC4', 'ITUB4', 'BBAS3', 'BBDC4'],
    'preco_fechamento': ['R$ 29,45', 'R$ 15,80', 'R$ 29,60', 'R$ 35,90', 'R$ 15,75'],
    'volume_negociado': ['55.430.200', '70.150.800', '49.800.100', '35.200.500', '68.950.000'],
    'variacao_dia': ['+0.52%', '-0.15%', '+0.80%', '+1.10%', '-0.32%']
}
df = pd.DataFrame(data)

print("--- 1. DataFrame Original do Pregão e seus Tipos (dtypes) ---")
df.info()


# 2. Preparação: Limpeza das colunas de texto antes da conversão.
# Esta etapa é crucial para o .astype() funcionar sem erros.

# Limpando a coluna de preço
df['preco_fechamento'] = df['preco_fechamento'].str.replace('R$ ', '', regex=False)
df['preco_fechamento'] = df['preco_fechamento'].str.replace(',', '.', regex=False)

# Limpando a coluna de volume (removendo separadores de milhar)
df['volume_negociado'] = df['volume_negociado'].str.replace('.', '', regex=False)

# Limpando a coluna de variação (removendo '%' e '+')
df['variacao_dia'] = df['variacao_dia'].str.replace('%', '', regex=False)
df['variacao_dia'] = df['variacao_dia'].str.replace('+', '', regex=False)

print("\n--- 2. DataFrame após a limpeza das colunas de texto ---")
print(df)


# 3. Agora, com os dados limpos, aplicamos o .astype() para as conversões corretas!
conversoes_financeiras = {
    'id_pregao': 'int',
    'data_pregao': 'datetime64[ns]',
    'ticker': 'category', # Ótimo para tickers que se repetem
    'preco_fechamento': 'float',
    'volume_negociado': 'int64', # Volume pode ser um número grande
    'variacao_dia': 'float'
}
df_corrigido = df.astype(conversoes_financeiras)

print("\n--- 3. DataFrame Corrigido e seus Novos Tipos ---")
df_corrigido.info()

# 4. Verificando a economia de memória, especialmente com 'ticker' como 'category'
memoria_antes = df.memory_usage(deep=True).sum()
memoria_depois = df_corrigido.memory_usage(deep=True).sum()

print(f"\n--- 4. Comparação de Memória ---")
print(f"Uso de memória ANTES da conversão: {memoria_antes} bytes")
print(f"Uso de memória DEPOIS da conversão: {memoria_depois} bytes")
print(f"Economia de {(1 - memoria_depois/memoria_antes):.2%}")

print("\n--- DataFrame Final, pronto para análise! ---")
print(df_corrigido)
```

```python
import pandas as pd

# 1. Criando nosso DataFrame com uma coluna de datas "problemática"
data = {
    'data_pregao_str': ['2025-10-06', '07/10/2025', '2025-Out-08', 
                        'Pregão Interrompido', '20251009'],
    'ibov_fechamento': [115200, 115950, 116500, 116550, 117100]
}
df = pd.DataFrame(data)
print("--- 1. DataFrame Original com Datas como Texto ---")
df.info()
print(df)


# 2. Tentativa de conversão direta (vai falhar por causa dos formatos mistos e do erro)
print("\n--- 2. Tentativa de conversão direta (esperamos um erro) ---")
try:
    pd.to_datetime(df['data_pregao_str'])
except Exception as e:
    print(f"Erro esperado: {e}")

# 3. A forma correta: Usando `errors='coerce'` para lidar com os problemas
print("\n--- 3. Convertendo com `errors='coerce'` ---")
df['data_pregao'] = pd.to_datetime(df['data_pregao_str'], dayfirst=True, errors='coerce')
df.info()
print(df)
print("\nRepare como 'Pregão Interrompido' virou NaT (Not a Time).")

# 4. O Poder do Acessor `.dt`: Extraindo informações da data
print("\n--- 4. Usando o acessor `.dt` para extrair informações ---")
df_final = df.dropna() # Removemos a linha com NaT para a análise
df_final['ano'] = df_final['data_pregao'].dt.year
df_final['mes'] = df_final['data_pregao'].dt.month
# Requer `locale` instalado
# Use the command locale -a on your terminal on Unix systems 
# to find your locale language code.
df_final['dia_da_semana'] = df_final['data_pregao'].dt.day_name(locale='pt_BR.utf8') 

print(df_final[['data_pregao', 'ano', 'mes', 'dia_da_semana']])
```

```python
import pandas as pd
import numpy as np

# 1. Criando nosso DataFrame com dados inconsistentes
data = {
    'ticker': ['PETR4', 'VALE3', 'MGLU3', 'BBDC4', 'ITUB4'],
    'corretora': ['Clear', 'Xp Investimentos', 'Rico', 'clear', 'XP'],
    'lado_operacao': [1, 1, -1, 1, -1], # 1 para Compra, -1 para Venda
    'setor': ['Petróleo', 'Mineração', 'Varejo', 'Financeiro', 'N/D'] # N/D = Não Definido
}
df = pd.DataFrame(data)

print("--- 1. DataFrame Original com Dados Inconsistentes ---")
print(df)


# 2. Usando LISTAS para padronizar as corretoras
# Queremos que todas as variações de 'Clear' e 'XP' sejam padronizadas.
print("\n--- 2. Padronizando as corretoras ---")
valores_antigos = ['Xp Investimentos', 'XP', 'clear']
valores_novos = ['XP', 'XP', 'Clear']
df_corrigido = df.replace(to_replace=valores_antigos, value=valores_novos)
print(df_corrigido)


# 3. Usando um DICIONÁRIO para recodificar e tratar "missing"
# Esta é a forma mais poderosa e legível!
print("\n--- 3. Recodificando 'lado_operacao' e tratando 'N/D' ---")
mapa_de_substituicao = {
    'lado_operacao': {1: 'Compra', -1: 'Venda'},
    'setor': {'N/D': np.nan}
}
# Note que aqui estamos usando um dicionário aninhado para especificar as colunas
df_corrigido = df_corrigido.replace(mapa_de_substituicao)
print(df_corrigido)


# 4. Verificando os tipos de dados após a substituição
print("\n--- 4. Info do DataFrame final ---")
df_corrigido.info()
# Agora podemos usar .fillna() na coluna 'setor' se quisermos!
```

```python
#gabarito 69
import numpy as np

serie_dados = pd.Series([20, 30, pd.NA, 25, 22, pd.NA, 17])

serie_dados = serie_dados.fillna(0)

retornos = []

for i, dado in enumerate(serie_dados):
    
    if i == 0:
        
        retornos.append(pd.NA)
    
    else:
        
        retorno = serie_dados[i]/serie_dados[i - 1] - 1
        
        retornos.append(retorno)
        
        
serie_retornos = pd.Series(retornos)

serie_retornos = serie_retornos.dropna()

serie_retornos = serie_retornos.replace([-1, np.inf], 0) #sempre pensem em listas quando precisar de mais de 1 elemento

serie_retornos
```
