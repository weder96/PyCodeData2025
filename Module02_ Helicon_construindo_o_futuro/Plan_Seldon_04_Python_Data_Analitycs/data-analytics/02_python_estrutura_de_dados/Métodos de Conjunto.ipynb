```python
#Criando um conjunto de números

a = {1, 2, 3}
b = {3, 4, 5}
c = {7, 8, 9}
d = {1, 2,  5, 8, 9}
```

```python
#Método Copy
```

```python
#Método união
print(a.union(b))
print(a | b)
```

```python
#Interseção
print(a.intersection(d))
print(a & d)
```

```python
#Diferença
print(c.difference(d))
print(c - d)

print(d.difference(c))
print(d - c)
```

```python
#Diferença simétrica
print(a.symmetric_difference(b))
```
