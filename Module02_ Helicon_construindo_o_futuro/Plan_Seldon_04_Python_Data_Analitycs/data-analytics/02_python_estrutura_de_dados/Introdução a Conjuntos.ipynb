**Conjuntos (Sets)**
Um set é uma coleção não ordenada, mutável e que não permite elementos duplicados.

```python
#Sintaxe de um set
frutas = {"banana", "laranja", "uva"}

print(frutas)
print(type(frutas))

#set vazio
frutas = {} #isso é um dicionário
frutas = set() #isso é um set vazio
```

```python
#Não é possível utilizar valores duplicados em sets
frutas = {"banana", "laranja", "uva", "banana"}

print(frutas)
```

```python
#Tamanho de um set
print(len(frutas))
```

```python
#Acessando um item de um set
for x in frutas:
  print(x)
```

```python
#Verificando se um item existe em um set

"banana" in frutas
```

```python
#Não é possível editar um elemento, mas é possível adicionar novos elementos

frutas.add("melancia")
print(frutas)
```

```python
#A função update também pode adicionar um elemento, mas ela itera sobre o que for adicionado

frutas.update("melancia")
print(frutas)
```

```python
#Concatenando sets

frutas1 = {"banana", "laranja", "uva"}
frutas2 = {"morango", "kiwi", "abacaxi"}

frutas1.update(frutas2)

print(frutas1)
```

```python
#Excluindo um valor de um set

#frutas1.remove("laranja")

print(frutas1)

#frutas1.discard("uva")

print(frutas1)
```

```python
#Verificando se duas listas são iguais
frutas1 = ["banana", "morango", "kiwi"]
frutas2 = ["morango", "morango", "banana", "morango", "kiwi", "kiwi", "banana", "morango"]

if set(frutas1) == set(frutas2):
  print("As frutas são iguais")
else:
  print("As frutas são diferentes")

```

```python
#Método clear

frutas1.clear()
print(frutas1)

#método del

del frutas1
print(frutas1)
```
