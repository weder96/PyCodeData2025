**F-Strings**
As f-strings são uma maneira de inserir variáveis e expressões diretamente dentro de uma string, de forma fácil e legível.

```python
#Formatação de Strings
nome = "Ana"
idade = 25
print(f"Meu nome é {nome} e tenho {idade} anos.")
```

```python
#Formatar números float

preco = 49.95823
print(f"O preço é R$ {preco:.2f}")
```

```python
#Formatar números inteiros (preenchimento com zeros)
numero = 7
print(f"O número é {numero:03}")
```

```python
#Alinhar texto: à esquerda, à direita e centralizado
texto = "Python"
print(f"{texto:<10} - alinhado à esquerda")
print(f"{texto:>10} - alinhado à direita")
print(f"{texto:^10} - centralizado")
```

```python
#Inserir separadores de milhar
valor = 1234567
print(f"Valor formatado: {valor:,}")
```

```python
#Exibir porcentagens
taxa = 0.085
print(f"Taxa de crescimento: {taxa:.2%}")
```

```python
#Representar números em diferentes bases (binário, hexadecimal, octal)
numero = 255
print(f"Binário: {numero:b}")
print(f"Octal: {numero:o}")
print(f"Hexadecimal: {numero:x}")
```

```python
#Cálculos e funções diretamente nas chaves
a = 5
b = 3
print(f"A soma de {a} + {b} é {a + b}")
```

```python
#Usando métodos de string
nome = "python"
print(f"Nome em maiúsculo: {nome.upper()}")
```

```python
#Apresentando o faturamento de uma empresa
empresa = "DataCorp"
faturamento = 1578934.58

print(f"A empresa {empresa} faturou R$ {faturamento:,.2f} neste mês.")
```
