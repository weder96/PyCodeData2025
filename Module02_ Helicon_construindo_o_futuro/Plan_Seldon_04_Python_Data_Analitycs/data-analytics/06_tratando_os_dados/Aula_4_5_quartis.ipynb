## üìä O que s√£o Quartis

Imagine uma fila de 100 pessoas ordenadas por altura.
Se voc√™ dividir essa fila em 4 partes iguais, cada parte ter√° 25 pessoas.

Quartis dividem os dados ordenados em quatro partes iguais:
- **Q1 (1¬∫ quartil)**: 25% dos dados est√£o abaixo
- **Q2 (Mediana)**: 50% dos dados est√£o abaixo
- **Q3 (3¬∫ quartil)**: 75% dos dados est√£o abaixo

**Objetivo:**

```python
import pandas as pd

# Carregar base de dados
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)

# Exibir primeiras linhas
df.head()
```

## üìê Calculando os Quartis
Vamos aplicar a fun√ß√£o `quantile()` no campo `total_bill`.

```python
quartis = df['total_bill'].quantile([0.25, 0.5, 0.75])
print("Quartis:")
print(quartis)
```

```python
quartis = df['tip'].quantile([0.25, 0.5, 0.75])
print("Quartis:")
print(quartis)
```

```python
df['total_bill'].mean()
```

## üì¶ Classificando por Quartis
Voc√™ tamb√©m pode usar `qcut()` para categorizar os dados em faixas baseadas nos quartis.

```python
df['segmento'] = pd.qcut(df['total_bill'], q=4, labels=['baixo', 'm√©dio-baixo', 'm√©dio-alto', 'alto'])
df[['total_bill', 'segmento']].head()
```
