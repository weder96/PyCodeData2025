**Estruturas Condicionais**
As estruturas condicionais são utilizadas para que o programa tome uma decisão com base em uma condição, que pode ser verdadeira ou falsa. Caso a condição seja verdadeira, ele realiza alguma instrução, caso seja falsa, ele realiza outra instrução.

```python
#Sintaxe do comando if

'''
if condição: (Se a condição for verdadeira)
  Executa esse bloco de comandos
'''
```

```python
#Exemplo com if
idade = 18

if idade >= 18:
  print('Maior de idade!')
```

```python
#Sintaxe do comando if-else

'''
if condição: (Se a condição for verdadeira)
  Executa esse bloco de comandos
else: (Se a condição do if for falsa)
  Executa esse bloco de comandos
'''
```

```python
#Exemplo com if-else

idade = 13

if idade >= 18:
  print('Maior de idade! ')
else:
  print('Menor de idade! ')
```

```python
#Sintaxe do comando if-elif-else

'''
if condição: (Se a condição for verdadeira)
  Executa esse bloco de comandos
elif condição: (Se a condição do if acima for falsa, ele faz uma nova verificação, se for verdadeira)
  Executa esse bloco de comandos
else: (Se a condição do if for falsa)
  Executa esse bloco de comandos
'''
```

```python
#Exemplo com if-elif-else
idade = 15

if idade < 18:
  print('Menor de idade! ')
elif idade < 66:
  print('Adulto! ')
else:
  print('Terceira idade!')
```

```python
#Exemplo utilizando operadores lógicos

idade = 22
tem_carteira = True

if idade >= 18 and tem_carteira:
  print("Pode dirigir.")
elif idade >= 18 and not tem_carteira:
  print('Não pode dirigir, pois não tem carteira')
```
