### Outros filtros no SQL (IN e LIKE)

```python
# Importando o pandas
import pandas as pd
```

```python
# Criando a conexão
import sqlite3
con = sqlite3.connect('BaseDados.db')
```

```python
# Criando um cursor
cur = con.cursor()
```

```python
# Criando uma função para consultar os dados
def executa_sql(comando):
    cur.execute(comando)
    resultado = cur.fetchall()
    resultado = pd.DataFrame(resultado)
    if resultado.shape[1] > 0:
        resultado.columns = [i[0] for i in cur.description]
    print(resultado.shape)
    display(resultado.head())
    return resultado
```

```python
# Utilizando parte do CASE que vimos na última aula
resultado_sql = executa_sql('SELECT * FROM DADOS')
```

```python
display(resultado_sql)
```

### IN
- Permite verificar se o valor está em uma lista de valores

```python
# Verificando os dias de último acesso iguais a 10,11,12 usando o WHERE + AND
resultado_sql = executa_sql('SELECT * FROM dados WHERE dias_ultimo_acesso >= 10 AND dias_ultimo_acesso <=12')
```

```python
# Alterando para usar o IN
resultado_sql = executa_sql('SELECT * FROM dados \
                            WHERE dias_ultimo_acesso IN (10,11,12)')
```

```python
# Utilizando o IN para filtrar as matrículas 28421,38273,42674,65749
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula FROM dados \
                            WHERE cod_matricula IN (28421,38273,42674,65749)')
```

### LIKE
- Permite verificar a existência de qualquer conjunto de caracteres na string

```python
# Verificando alunos que o nome seja igual Maria Ferreira utilizando o IN
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula FROM dados \
                            WHERE nome_aluno IN ("Maria Ferreira")')
```

```python
# Fazendo a mesma consulta utilizando o LIKE
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula FROM dados \
                            WHERE nome_aluno LIKE "Maria Ferreira"')
```

```python
# Verificando todas as alunas chamadas Maria
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula FROM dados \
                            WHERE nome_aluno LIKE "Maria%"')
```

```python
# Verificando se o e-mail possui o caracter !
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula,[e-mail],acesso_plataforma,acesso_liberado FROM dados \
                            WHERE [e-mail] LIKE "%!%"')
```

```python
# Verificando se possui !,ã ou í
resultado_sql = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula,[e-mail],acesso_plataforma,acesso_liberado FROM dados \
                            WHERE [e-mail] LIKE "%!%" \
                            OR [e-mail] LIKE "%ã%" \
                            OR [e-mail] LIKE "%í%"')
```
