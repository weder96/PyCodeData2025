**TUPLAS**
Uma tupla é uma estrutura de dados utilizada também para armazenar mais de um dado por vez na mesma estrutura. Porém, ela tem características diferentes de uma lista.
- Tuplas são imutáveis, ou seja, não é possível alterar um elemento que faz parte dela.

```python
#Definindo uma tupla (SINTAXE)
carros = ('Gol', 'Voyage', 'Fusca')
print(carros)
```

```python
#Verificando o tipo de dados
print(type(carros))
```

```python
#Acessando um elemento de uma tupla
print(carros[0])
print(carros[1])
print(carros[-1])
```

```python
#Tuplas são imutáveis
carros[0] = 'Hilux'
```

```python
#Alterando dados de uma tupla
frutas = ("abacaxi", "banana", "limão", "laranja", "maracujá", "maçã")
print(frutas)
auxiliar = list(frutas)
auxiliar[1] = "kiwi"
frutas = tuple(auxiliar)

print(frutas)
```

```python
#Posso ter elementos iguais dentro de uma tupla
frutas = ('Abacaxi', 'Maçã', 'Limão', 'Abacaxi')
print(frutas)
```

```python
#Verificando o tamanho de uma tupla
print(len(carros))
print(len(frutas))
```

```python
#Extraindo dados de uma tupla em variáveis
numeros = (10, 50, 14)

(n1, n2, n3) = numeros

print(f"O primeiro valor é {n1}, o segundo valor é {n2} e o terceiro valor é {n3}")

#Se você não sabe a quantidade de elemento, pode usar o *

numeros = (10, 50, 14, 9, 237, 80, 41)

(n1, n2, *n3) = numeros
```

```python
#Percorrando uma tupla
frutas = ('Abacaxi', 'Maçã', 'Limão', 'Goiaba')

for x in frutas:
  print(x)
```

```python
#Juntando tuplas
frutas1 = ('Morango', 'kiwi')
frutas2 = ('Maçã', 'Banana')

frutas3 = frutas1 + frutas2

print(frutas3)
```

```python
#count
frutas = ('Abacaxi', 'Maçã', 'Limão', 'Abacaxi')

frutas.count('Abacaxi')
```

```python
#index

frutas = ('Abacaxi', 'Maçã', 'Limão', 'Abacaxi')

frutas.index('Limão')
```
