```python
#Tamanho de uma lista
cores = ['azul', 'amarelo', 'vermelho']
print(len(cores))
```

```python
#Inserindo dados em uma lista (append)
cores = ['azul', 'amarelo', 'vermelho']

cores.append('lilás')
print(cores)
cores.append('marrom')
print(cores)
```

```python
#Inserindo dados em uma posição específica (insert)
frutas = ['laranja', 'goiaba', 'abacate']
print(frutas)

frutas.insert(1, 'maracujá')
print(frutas)

frutas.insert(0, 'Jaboticaba')
print(frutas)
```

```python
#Juntando listas com o extend
cores1 = ['azul', 'amarelo']
cores2 = ['vermelho', 'laranja']
print(cores1)

cores1.extend(cores2)

print(cores1)
print(cores2)
```

```python
#Removendo dados de uma lista (remove)
frutas = ['laranja', 'goiaba', 'abacate']

print(frutas)

frutas.remove('goiaba')

print(frutas)
```

```python
#Removendo dados de uma lista (pop)
frutas = ['laranja', 'goiaba', 'abacate']
print(frutas)

frutas.pop()

print(frutas)

frutas.pop(0)

print(frutas)
```

```python
#Removendo de uma lista (del)
frutas = ['laranja', 'goiaba', 'abacate']
print(frutas)

del frutas[0]
print(frutas)
```

```python
#Excluindo uma lista(del)
frutas = ['laranja', 'goiaba', 'abacate']
print(frutas)

del frutas

print(frutas)
```

```python
#Limpando todos os itens de uma lista(clear)
frutas = ['laranja', 'goiaba', 'abacate']

print(frutas)

frutas.clear()

print(frutas)
```
