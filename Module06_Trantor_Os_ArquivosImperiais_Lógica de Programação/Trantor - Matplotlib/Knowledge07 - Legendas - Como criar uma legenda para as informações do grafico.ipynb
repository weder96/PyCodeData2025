# Mundo 7 - Como criar legendas em um gr√°fico

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)
```

```python
cotacoes = yf.download(["WEGE3.SA", "PETR4.SA", "VALE3.SA", "LREN3.SA"])['Adj Close']

fig, ax = plt.subplots()

for empresa in cotacoes.columns:
      
    ax.plot(cotacoes.index, cotacoes[empresa].values, label = empresa)

ax.legend(loc = 10) 
```

```python
#gabarito 96

acao1 = str(input("Escolha a 1¬∫ a√ß√£o: "))
acao2 = str(input("Escolha a 2¬∫ a√ß√£o: "))
acao3 = str(input("Escolha a 3¬∫ a√ß√£o: "))
acao4 = str(input("Escolha a 4¬∫ a√ß√£o: "))

lista_acoes = [acao1, acao2, acao3, acao4]

lista_acoes = [acao + ".SA" for acao in lista_acoes]

base_dados = yf.download(lista_acoes)['Adj Close']

fig, ax = plt.subplots()

for empresa in base_dados.columns:
      
    ax.plot(base_dados.index, base_dados[empresa].values, label = empresa.upper())

ax.legend(loc = 0)
```
