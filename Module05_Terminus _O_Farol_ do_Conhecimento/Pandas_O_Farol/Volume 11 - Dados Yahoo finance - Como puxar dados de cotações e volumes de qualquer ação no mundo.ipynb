```python
#puxar ddos várias ações
acao = ["META", "NKE", "GOOGL", "AMZN"]
dados = yf.download(tickers= acao, start = "2024-12-31", end = datetime.now(),  auto_adjust=False)                         
dados
```

```python
# Pega a coluna 'Close', que retorna um novo DataFrame com todas as ações
precos_fechamento = dados['Close']

print("--- Preços de Fechamento de TODAS as Ações ---")
print(precos_fechamento.head())
```

```python
# Pega todos os dados da ação 'META'
dados_meta = dados.xs('META', level=1, axis=1)

print("\n--- Todos os Dados da Ação META ---")
print(dados_meta.head())
```
