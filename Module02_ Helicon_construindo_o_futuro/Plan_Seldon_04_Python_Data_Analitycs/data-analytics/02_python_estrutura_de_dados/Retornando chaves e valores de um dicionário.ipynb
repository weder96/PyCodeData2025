```python
#Acessando valores de um dicionário (indice, get)
pessoas = {
    "nome": "Sonia dos Santos",
    "idade": 32,
    "altura": 1.69,
    "habilitado": False
}

print(pessoas["altura"])

print(pessoas.get("altura"))
```

```python
#Retornando as chaves de um dicionário

print(pessoas.keys())
```

```python
#Retornando os valores de um dicionário

print(pessoas.values())
```

```python
#Retornando chave e valor

print(pessoas.items())
```

```python
#Verificando a existência de uma chave específica

"nome" in pessoas
```

```python
#Alterando valores no dicionário

pessoas["idade"] = 40
pessoas["habilitado"] = True

print(pessoas)

#Com update

pessoas = {
    "nome": "Sonia dos Santos",
    "idade": 32,
    "altura": 1.69,
    "habilitado": False
}

print(pessoas)

pessoas.update({"idade" : 40, "habilitado": True})

print(pessoas)
```

```python
#Adicionando um novo chave:valor no dicionário

pessoas["cpf"] = "000.000.000-00"

print(pessoas)

pessoas.update({"rg" : "55.555.555-5"})

print(pessoas)
```

```python
#Removendo um item chave:valor no dicionário

#pessoas.pop("rg")

print(pessoas)

```

```python
#Removendo o último elemento inserido
pessoas.popitem()

print(pessoas)
```

```python
#Utilizando o del

cliente = {
    "nome" : "Priscila Costa",
    "data_nasc" : "28-04-1996",
    "cpf" : "000.000.000-000",
    "rg" : "00.000.000-0"
}

print(cliente)

del cliente["cpf"]

print(cliente)

#del cliente
#print(cliente)

#Clear

cliente.clear()

print(cliente)
```
