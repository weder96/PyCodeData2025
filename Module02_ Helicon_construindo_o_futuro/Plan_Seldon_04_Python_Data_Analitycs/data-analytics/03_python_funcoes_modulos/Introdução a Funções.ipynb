#**FUNÇÕES**
Funções são blocos de código que executam uma tarefa específica. Elas são definidas uma vez e podem ser reutilizadas sempre que necessário.

```python
#Sintaxe de uma função
def nome_da_funcao():
  print()
  #bloco de códigos da função
```

```python
#Exemplo de uma função que imprime na tela um texto
def imprime_texto():
    print("Módulo de funções em Python")
```

```python
#Chamando a função
imprime_texto()
```

```python
#Utilizando variáveis com funções
def soma():
  n1 = int(input("Digite o primeiro número: "))
  n2 = int(input("Digite o segundo número: "))
  soma = n1 + n2
  print(f"{n1} + {n2} = {soma}")
```

```python
#soma
soma()
```

```python
#Posso chamar a função quantas vezes for necessário
soma()
```

##**Parâmetros**
Você pode enviar valores para as funções através dos parâmetros. Esses valores são enviados dentro dos parênteses das funções. Você pode enviar um ou mais valores, desde que separados por vírgula.


```python
#Sintaxe de uma função com parâmetros
def funcao_com_parametros(nome_parametro):
  print(f"{nome_parametro}")
```

```python
#Chamar uma função com parâmetro
funcao_com_parametros("Aqui eu envio o(s) valor(es)")
```

```python
#Chamar uma função com argumento (Envio de valores de qualquer tipo)
funcao_com_parametros(10)
funcao_com_parametros(True)
funcao_com_parametros([50, 30, 60])
funcao_com_parametros(10, 50, 20) #Erro, pois a função espera apenas um parâmetro e estão sendo enviados 3
```

```python
#Utilizando o *args - Permite o envio de n valores como parâmetros, caso você não saiba a quantidade
def funcao_com_parametros(*valores):
  print(valores)
```

```python
funcao_com_parametros(3,12,4,7)
```

```python
#Função com parâmetros nomeados com quantidade definida - key -> args
def funcao_com_key_args(idade, nome, altura):
  print(f"A pessoa de nome {nome} possui altura {altura} e {idade} anos de idade" )

nome = input("Digite o seu nome: ")
idade = int(input("Digite a sua idade: "))
altura = float(input("Digite a sua altura: "))
#funcao_com_key_args(nome, idade, altura)
funcao_com_key_args(nome = nome, idade = idade, altura = altura)
```

```python
#Funções com parâmetros nomeados sem quantidade definida - **kwargs
def funcao_com_key_args(**valores):
  print(valores["nome"])
  print(valores["idade"])
  for x in valores.keys():
    print(valores[x])

  for y in valores.values():
    print(f"{y}")

  for x, y in valores.items():
    print(f"{x}: {y}")

funcao_com_key_args(nome = "Osvaldo", idade = "33")
```

##**Valores Padrão para os parâmetros**

```python
#Utilizando um valor default
def funcao_com_parametro_padrao(cidade = 'São Paulo'):
  print(cidade)


#cidade = input("Digite o nome da sua cidade: ")
funcao_com_parametro_padrao('Rio de Janeiro')
funcao_com_parametro_padrao('Fortaleza')
funcao_com_parametro_padrao()
```

##**Estruturas de Dados com Funções**

```python
#Utilizando parâmetros de funções com lista
def imprimir_lista_de_frutas(frutas):
  print(frutas)
  print(type(frutas))

  for x in frutas:
    print(x)

lista_frutas = ["banana", "laranja", "abacaxi", "maracujá"]
imprimir_lista_de_frutas(lista_frutas)
```

```python
#Utilizando parâmetros de funções com dicionários
def imprimir_clientes(clientes):
  print(clientes)
  print(type(clientes))

  for x, y in clientes.items():
    print(f"{x}: {y}")

cliente = {"nome": "Osvaldo da Silva", "idade" : 26, "cpf": "000.000.000-00"}
imprimir_clientes(cliente)
```

##**Funções com retorno**

```python
#Função retornando valores inteiros
def funcao_com_retorno(valor):
  return valor ** 2

numero = funcao_com_retorno(10)

print(numero)
```

```python
#Função retornando valores de outros tipos
def funcao_com_retorno(texto):
  return texto
retorno = funcao_com_retorno("Python - Funções")

print(retorno)
```

```python
#Função retornando estruturas de dados - lista
def funcao_com_retorno():
  return ["banana", "limão", "abacate", "uva"]

lista_frutas = funcao_com_retorno()

print(lista_frutas)
print(type(lista_frutas))

for x in lista_frutas:
  print(x)
```

```python
#Função retornando estruturas de dados - Dicionário
def funcao_com_retorno():
  return {"nome":"Osvaldo", "idade":32, "cpf":"000.000.000-00"}

cliente = funcao_com_retorno()

print(cliente)
print(type(cliente))

for x, y in cliente.items():
  print(f"{x}:{y}")
```

```python
#Bloqueando a utilização de parâmetros nomeados
def funcao_sem_parametros_nomeados(x, /):
  print(x)

funcao_sem_parametros_nomeados(10)
funcao_sem_parametros_nomeados(x = 3)
```

```python
#obrigando a utilização de parâmetros nomeados
def funcao_com_parametros_nomeados(*, x):
  print(x)

funcao_com_parametros_nomeados(x = 4)
funcao_com_parametros_nomeados(4)
```

```python
#Utilizando parâmetros nomeados e não nomeados
def funcao_com_e_sem_parametros_nomeados(x, y, /, *, z, nome, idade):
  print(x)
  print(y)
  print(z)
  print(nome)
  print(idade)

funcao_com_e_sem_parametros_nomeados(10, 1.88, z = 10, nome = "Priscila", idade = 8)
```
