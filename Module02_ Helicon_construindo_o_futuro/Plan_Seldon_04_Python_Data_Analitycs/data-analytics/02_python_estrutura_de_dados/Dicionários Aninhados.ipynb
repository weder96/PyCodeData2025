```python
#Criando um dicionário aninhado

clientes = {
    "cliente1": {
        "nome": "Ana Paula",
        "idade": 32,
        "email": "ana.paula@email.com",
        "cidade": "São Paulo"
    },
    "cliente2": {
        "nome": "Roberto Lima",
        "idade": 45,
        "email": "roberto.lima@email.com",
        "cidade": "Rio de Janeiro"
    },
    "cliente3": {
        "nome": "Mariana Rocha",
        "idade": 28,
        "email": "mariana.rocha@email.com",
        "cidade": "Curitiba"
    }
}
print(clientes)
```

```python
#Acessando elementos de um dicionário aninhado
print(clientes["cliente1"]["idade"])
```

```python
#Percorrendo um dicionário aninhado

for chave, valor in clientes.items():
  print(chave)
  for chave2, valor2 in valor.items():
    print(f"{chave2} : {valor2}")
```

```python
#Adicionando elementos em um dicionário aninhado

clientes["cliente4"] = {
    "nome": "Mariana Rocha",
    "idade": 28,
    "cidade": "Curitiba"
}
```

```python
print(clientes)
```

```python
#Atualizando os dados de um dicionário aninhado
clientes["cliente1"]["idade"] = 33
```

```python
#Removendo um elemento de um dicionário aninhado

del clientes["cliente2"]
```
